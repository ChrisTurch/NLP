```python
import pandas as pd 
import re 
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer() 
```

```python
data = pd.read_csv('cyberbullying_tweets.csv')
data.head()
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

```python
count_vect = CountVectorizer(analyzer=clean_text)
X_counts = count_vect.fit_transform(data['tweet_text'])
```

```python
print(X_counts.shape)
```

```python
data_sample = data[0:35]

count_vect_sample = CountVectorizer(analyzer=clean_text)
X_counts_sample = count_vect_sample.fit_transform(data_sample['tweet_text'])
```

```python
print(X_counts_sample.shape)
```

```python
print(count_vect_sample.get_feature_names_out())
```

```python
X_counts_df = pd.DataFrame(X_counts_sample.toarray())
X_counts_df
```

```python
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?',
 ]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names_out()
```

```python
def clean_text2(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join(ps.stem(word) for word in tokens if word not in stopwords)
    return text
```

```python
data['cleaned_text'] = data['tweet_text'].apply(lambda x: clean_text2(x))
data.head()
```

```python
data_sample = data[0:35]

ngram_vect_sample = CountVectorizer(ngram_range=(2, 2))
X_ngram_sample = ngram_vect_sample.fit_transform(data_sample['cleaned_text'])
```

```python
print(ngram_vect_sample.get_feature_names_out())
```

```python
ngram_vect_sample2 = CountVectorizer(analyzer='word', ngram_range=(2, 2))
X_ngram_sample2 = ngram_vect_sample2.fit_transform(data_sample['tweet_text'])
```

```python
print(ngram_vect_sample2.get_feature_names_out())
```
